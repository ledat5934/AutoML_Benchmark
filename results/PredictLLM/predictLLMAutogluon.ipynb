```python
from sklearn.model_selection import train_test_split
import pandas as pd
train_df = pd.read_csv("train.csv")
train_data, valid_data = train_test_split(train_df, test_size=0.2, random_state=42, stratify=train_df['target'])
```

```python
from autogluon.tabular import TabularPredictor

predictor = TabularPredictor(
    label="target",
    problem_type="multiclass",
    eval_metric="accuracy",
    path="ag_model"
)

predictor.fit(
    train_data=train_data,
    time_limit=300
)
```

```python
predictor.leaderboard(valid_data, extra_metrics = ['accuracy', 'log_loss', 'f1_macro', 'f1_weighted', 'roc_auc_ovr', 'roc_auc_ovr_weighted'])
```
